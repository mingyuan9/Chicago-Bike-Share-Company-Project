```python
from peewee import *
from datetime import datetime
```

```python
db = SqliteDatabase('database.db')
```

```python
class BaseModel(Model):
    class Meta:
        database = db
        
class BikeModel(BaseModel):
    class Meta:
        table_name = 'bike_data'
        
    ride_id = TextField(null=True)
    rideable_type = TextField(null=True)
    started_at = DateTimeField(null=True)
    ended_at = DateTimeField(null=True)
    start_station_name = TextField(null=True)
    start_station_id = TextField(null=True)
    end_station_name = TextField(null=True)
    end_station_id = TextField(null=True)
    start_lat = FloatField(null=True)
    start_lng = FloatField(null=True)
    end_lat = FloatField(null=True)
    end_lng = FloatField(null=True)
    member_casual = TextField(null=True)
```

```python
db.create_tables([BikeModel])
dtfmt = "%Y-%m-%d %H:%M:%S" # 2023-01-22 11:50:20
```

```python
def to_float(string):
    try:
        res = float(string)
        return res
    except ValueError:
        return None
    
    
for id in range(0,6):
    print(f"/Users/mingyuan/Desktop/Career/Projects/Cousera - Google Data Analytics/Case Study/20230{id+1}-divvy-tripdata.csv")
    ##print(f"writing 20230{id+1}-divvy-tripdata.csv")

    ##with open(f'20230{id+1}-divvy-tripdata.csv', 'r') as f:
    with open(f'/Users/mingyuan/Desktop/Career/Projects/Cousera - Google Data Analytics/Case Study/20230{id+1}-divvy-tripdata.csv','r') as f:
        reader = csv.reader(f)
        header = next(reader)
        for r in reader:
            try:
                BikeModel.create(
                    ride_id = r[0],
                    rideable_type = r[1],
                    started_at = datetime.strptime(r[2], dtfmt),
                    ended_at = datetime.strptime(r[3], dtfmt),
                    start_station_name = r[4],
                    start_station_id = r[5],
                    end_station_name = r[6],
                    end_station_id = r[7],
                    start_lat = to_float(r[8]),
                    start_lng = to_float(r[9]),
                    end_lat = to_float(r[10]),
                    end_lng = to_float(r[11]),
                    member_casual = r[12],
                )
            except IntegrityError:
                print(f"failed processing {r}")
#             break
#     break
```
